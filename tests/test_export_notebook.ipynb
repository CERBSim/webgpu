```python
import numpy as np
from webgpu.jupyter import Draw
from webgpu.triangles import TriangulationRenderer

points = np.array([[-0.5, -0.5, 0.0], [0.5, -0.5, 0.0], [0.0, 0.5, 0.0]], dtype=np.float32)
scene = Draw(TriangulationRenderer(points, color=(0.0, 1.0, 0.0, 1.0)))
```
